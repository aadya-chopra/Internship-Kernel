```python
# Step 2: Data Cleaning and Preprocessing

if iris.isnull().sum().sum() > 0:
    iris.fillna(method='ffill', inplace=True) #filling missing values if any

# changing 'species'(categorical) into numeric values
iris['species_encoded'] = iris['species'].astype('category').cat.codes
iris['species_encoded'] = pd.to_numeric(iris['species_encoded'])
```

```python
#histogram plots
iris.hist(bins=15, figsize=(10, 6), edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16)
plt.show()
```

```python
#pairplotting of dataset
sns.pairplot(iris, hue='species', palette='Set2')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
#box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris.drop(columns=['species_encoded']), orient='h', palette='Set3')
plt.title('Box Plots of Iris Dataset Features')
plt.show()
```

```python
columns_to_convert = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] #conversion into float
iris[columns_to_convert] = iris[columns_to_convert].astype(float)
```
